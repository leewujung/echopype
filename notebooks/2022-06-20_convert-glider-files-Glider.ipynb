# Convert all files from 2021 glider fieldwork -- ADCP on glider

```python
from pathlib import Path
import numpy as np
import xarray as xr

import echopype as ep
```

## First deployment: recovered 2021-08-29

```python
data_path = Path("/Volumes/MURI_4TB/OER_glider_ADCP/ADCP_on_glider/20210829_recover/")
save_path = Path("/Volumes/MURI_4TB/OER_glider_ADCP/ADCP_on_glider_converted/20210829_recover/")
```

```python
if not save_path.exists():
    save_path.mkdir(exist_ok=True)
```

```python
files = list(data_path.glob("*.ad2cp"))
```

```python
files
```

```python
val_error_file_list_deploy1 = []
for ff in files:
    try:
        ed = ep.open_raw(ff, sonar_model="AD2CP")
        ed.to_zarr(save_path, overwrite=True)
    except:
        val_error_file_list_deploy1.append(ff)
```

```python
val_error_file_list_deploy1
```
